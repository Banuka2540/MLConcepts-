```python
# importing libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# read the data file 
data = pd.read_csv('breast_cancer_bd.csv')
data.sample(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Bare Nuclei'].values
```

```python
data['Bare Nuclei'].unique()
```

```python
# dropping the ? mark
data['Bare Nuclei'] = data['Bare Nuclei'].replace('?',np.nan)
```

```python
data['Bare Nuclei'] = pd.to_numeric(data['Bare Nuclei'])
```

```python
data.info()
```

```python
# change the missing values into median value and fill them 
data['Bare Nuclei'] = data['Bare Nuclei'].fillna(data['Bare Nuclei'].median())
```

```python
# plot the data 

plt.bar(data['Clump Thickness'],data['Class'])
plt.show()
```
