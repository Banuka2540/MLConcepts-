```python
# importing libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# read the data file 
data = pd.read_csv('breast_cancer_bd.csv')
data.sample(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Bare Nuclei'].values
```

```python
data['Bare Nuclei'].unique()
```

```python
# dropping the ? mark
data['Bare Nuclei'] = data['Bare Nuclei'].replace('?',np.nan)
```

```python
data['Bare Nuclei'] = pd.to_numeric(data['Bare Nuclei'])
```

```python
data.info()
```

```python
# change the missing values into median value and fill them 
data['Bare Nuclei'] = data['Bare Nuclei'].fillna(data['Bare Nuclei'].median())
```

```python
data['Bare Nuclei']
```

```python
# plot the data 

fig,ax = plt.subplots(3,3,figsize = (14,14))

sns.scatterplot(x='Clump Thickness',y='Class', data=data, ax=ax[0,0])
ax[0,0].set_title('Clump Thickness vs Class')

sns.barplot(x='Uniformity of Cell Size',y='Class', data=data, ax=ax[0,1])
ax[0,1].set_title('Uniformity of Cell Size vs Class')

sns.boxplot(x='Uniformity of Cell Shape',y='Class', data=data, ax=ax[0,2])
ax[0,2].set_title('Uniformity of Cell Shape vs Class')

sns.boxplot(x='Marginal Adhesion',y='Class', data=data, ax=ax[1,0])
ax[1,0].set_title('Marginal Adhesion vs Class')

sns.boxplot(x='Single Epithelial Cell Size',y='Class', data=data, ax=ax[1,1])
ax[1,1].set_title('Single Epithelial Cell Size vs Class')

sns.boxplot(x='Bare Nuclei',y='Class', data=data, ax=ax[1,2])
ax[1,2].set_title('Bare Nuclei vs Class')

sns.boxplot(x='Bland Chromatin',y='Class', data=data, ax=ax[2,0])
ax[2,0].set_title('Bland Chromatin vs Class')

sns.boxplot(x='Normal Nucleoli',y='Class', data=data, ax=ax[2,1])
ax[2,1].set_title('Normal Nucleoli vs Class')

sns.scatterplot(x='Mitoses',y='Class', data=data, ax=ax[2,2])
ax[2,2].set_title('Mitoses vs Class')

plt.show()
```

```python
# divide the dataset into train test split 
from sklearn.model_selection import train_test_split
```

```python
# divide the features 
x = data.drop(['Class', 'Sample code number'], axis=1)
y = data['Class']
```

```python
x.info()
```

```python
y.info()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
results = pd.DataFrame({
    'Predicted' : y_pred,
    'Actual' : y_test
})
print(results)
```

```python
from sklearn.metrics import accuracy_score
     
accuracy = accuracy_score(y_test, y_pred) 
print("Accuracy:", accuracy)
```
