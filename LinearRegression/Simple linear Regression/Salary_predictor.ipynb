```python
# importing the libraries 
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
salary_data = pd.read_csv('salary_dataset.csv')
salary_data.head()
```

```python
salary_data.info()
```

```python
salary_data = salary_data[['YearsExperience','Salary']]
```

```python
salary_data.sample(5)
```

```python
# plotting the columns 

figure = figsize = (3,4)
plt.bar(salary_data.YearsExperience,salary_data.Salary)
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```

```python
# divide the dataset into train test split 
from sklearn.model_selection import train_test_split
```

```python
x = salary_data[['YearsExperience']]
y = salary_data['Salary']

x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# creating the model for predict
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
results = pd.DataFrame({
    'years_of_experience' :x_test['YearsExperience'].values,
    'Actual_Salary' : y_test.values,
    'Predicted_Salary' : y_pred
})
print(results)
```

```python
# check the accuracy 
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error 
```

```python
r2 = r2_score(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
print(r2)
print(mae)
print(mse)
print(rmse)
```
